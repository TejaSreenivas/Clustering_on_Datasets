#### AUDIO

```python
cl = 41
mx = 51
plt.figure(1)
plt.gcf().set_size_inches(10, 10)
plt.subplot(221)
plt.plot(list(range(1,mx)),k_au['sse'].values(),label='kmeans',color='r')
plt.plot(list(range(1,mx)),g_au['sse'].values(),label='GMM',color='b')
plt.plot(list(range(1,mx)),h_au['sse'].values(),label='Agg',color='g')
plt.scatter(cl,k_au['sse_c'],color='r')
#plt.scatter(cl,g_au['sse_c'],color='b')
#plt.scatter(cl,h_au['sse_c'],color='g')
plt.xlabel('N')
plt.ylabel('SSE')
plt.grid(True)
plt.title('Elbow')


plt.subplot(222)
plt.plot(list(range(2,mx)),k_au['silh'].values(),label='kmeans',color='r')
plt.plot(list(range(2,mx)),g_au['silh'].values(),label='GMM',color='b')
plt.plot(list(range(2,mx)),h_au['silh'].values(),label='Agg',color='g')
plt.scatter(cl,k_au['silh_c'],color='r')
#plt.scatter(cl,g_au['silh_c'],color='b')
#plt.scatter(cl,h_au['silh_c'],color='g')
plt.xlabel('N')
plt.ylabel('silhouette coefficient')
plt.grid(True)
plt.title('silhouette')

plt.subplot(223)
plt.plot(list(range(1,mx)),k_au['gap'],label='kmeans',color='r')
plt.plot(list(range(1,mx)),g_au['gap'],label='GMM',color='b')
plt.plot(list(range(1,mx)),h_au['gap'],label='Agg',color='g')
plt.scatter(cl,k_au['gap_c'],color='r')
#plt.scatter(cl,g_au['gap_c'],color='b')
#plt.scatter(cl,h_au['gap_c'],color='g')
plt.xlabel('N')
plt.ylabel('GAP value')
plt.grid(True)
plt.title('GAP statistic')

plt.subplot(224)
plt.plot(list(range(2,mx)),k_au['chi'],label='kmeans',color='r')
plt.plot(list(range(2,mx)),g_au['chi'],label='GMM',color='b')
plt.plot(list(range(2,mx)),h_au['chi'],label='Agg',color='g')
plt.scatter(cl,k_au['chi_c'],color='r')
#plt.scatter(cl,g_au['chi_c'],color='b')
#plt.scatter(cl,h_au['chi_c'],color='g')
plt.xlabel('N')
plt.ylabel('CH value')
plt.grid(True)
plt.title('CH coefficient')


plt.legend()
# Format the minor tick labels of the y-axis into empty strings with
# `NullFormatter`, to avoid cumbering the axis with too many labels.
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=.55,
                    wspace=0.35)

plt.show()
fig1 = plt.gcf()
plt.show()
plt.draw()
fig1.savefig('audio.png', dpi=300,bbox_inches="tight")
```
