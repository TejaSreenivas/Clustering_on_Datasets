```python
import numpy as np
import os
import csv
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn import datasets
import seaborn as sns
import pandas as pd
```

### IRIS dataset

```python
iris = datasets.load_iris()
```

```python
x = StandardScaler().fit_transform(iris.data)
x_reduced = PCA(n_components=2).fit_transform(x)
labels = iris.target
```

```python
PCA(n_components=2).fit(x).explained_variance_ratio_.sum()*100
```

#### plot 

```python
print("Number of classes:  {}".format(labels.max()+1))
print("Dataset shape  :",format(x.shape))
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('iris dataset', fontsize = 20)
targets = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colors = ['r', 'g', 'b']

ax.scatter(x_reduced[:,0]
            , x_reduced[:,1]
            , c = np.array(colors)[:][labels]
            , s = 50, edgecolor='none')
ax.grid()
```

### Digits dataset 

```python
from sklearn.datasets import load_digits
digits = load_digits()
labels = digits.target
x = digits.data
```

```python
x = StandardScaler().fit_transform(x)
x_reduced = PCA(n_components=2).fit_transform(x)
print(PCA(n_components=2).fit(x).explained_variance_ratio_.sum()*100)
print("Number of classes:  {}".format(labels.max()+1))
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('digits', fontsize = 20)

ax.scatter(x_reduced[:,0]
            , x_reduced[:,1]
            , c = labels
            , s = 50, edgecolor='none')
ax.grid()
```

### Glass dataset 

```python
import csv
with open(dep_path+'/clustering/glass.csv', 'r') as f:
    yeast = list(csv.reader(f, delimiter=','))
x = np.array(yeast)
x = x[1:,:-2].astype(np.float32)
labels = np.array(yeast)[1:,-1].astype(np.int32)
```

```python
x = StandardScaler().fit_transform(x)
x_reduced = PCA(n_components=2).fit_transform(x)
print(PCA(n_components=2).fit(x).explained_variance_ratio_.sum()*100)
print("Number of classes:  {}".format(labels.max()+1))
print("Dataset shape  :",format(x.shape))
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('Glass', fontsize = 20)

ax.scatter(x_reduced[:,0]
            , x_reduced[:,1]
            , c = labels
            , s = 50, edgecolor='none')
ax.grid()
```
